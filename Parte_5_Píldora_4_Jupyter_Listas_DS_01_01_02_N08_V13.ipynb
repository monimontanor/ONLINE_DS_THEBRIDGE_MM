## Listas
[al indice](#Contenidos)  
Se trata de otro de los tipos de datos de Python más usados. Dentro de las colecciones, que veremos más adelante, la lista es la colección que normalmente se le da más uso. **Nos permiten almacenar conjuntos de variables u objetos**, y son elementos de lo más versátiles puesto que podemos almacenar objetos de distintos tipos, modificarlos, eliminarlos, meter listas dentro de listas... Sus dos caractrísticas principales son:
* **Mutables**: una vez se ha creado la lista, se puede modificar
* **Ordenada**: Los elementos tienen un cierto orden, lo que nos permite acceder al elemento que queramos teniendo en cuenta tal orden

En cuanto a su sintaxis, cuando declaremos la lista simplemente hay que separar cada elemento con comas, y rodearlo todo con corchetes.

```python
# Listas de números

numeros = [1,2,3,4,5]
print(numeros, type(numeros))
```

```python
# Lista de strings

strings = ["Esto es una cadena de strings", "Esto", "es", "una", "prueba"]
print(strings, type(strings))

# Lista de booleanos

booleanos = [True, False, True or False, False and True, False and False, True or True, True and False, False and False, False and True]
print(booleanos, type(booleanos))
```

```python
# Listas de elementos mezclados

mix = ["Esto es una mezcla de todo", 24, -16, False and True, False, True or False]
print(mix)
type(mix)            
```

```python
# Lista anidada

list_nido = [["Esto es una mezcla", 24, -16, False and True, False, True or False], "Y aqui seguimos", True and False]
print(list_nido)
type(list_nido)
```

```python
len(mix)
```

```python
len(list_nido)
```

```python
concatenacion = mix + list_nido
print(concatenacion)
print(type(concatenacion))
```

```python
# Prueba cambiando el orden de la suma para ver como cambia la posición
concatenacion2 = list_nido + mix
print(concatenacion2)
print(type(concatenacion2))
```

Accedemos a los elemenos de la lista mediante corchetes `[]`

**Importante**. El primer elemento es el 0

```python
#Explicación; comencé con una lista anidada, por eso el primer objeto (es decir el 0) es el de la lista anidada

list_ejemplo = [["Esto es una mezcla", 24, -16, False and True, False, True or False], "Y aqui seguimos", True and False]
print(list_ejemplo[0])
```

```python
list_ejemplo = [["Esto es una mezcla", 24, -16, False and True, False, True or False], "Y aqui seguimos", True and False]
print(list_ejemplo[0])
print(list_ejemplo[-1])
print(list_ejemplo[-2])
print(list_ejemplo[3])

#Que bello ejemplo!. El último es True and False, por lo tanto el último (-1) es False
```

### Metodos en Listas
Para el tipo de objeto lista, también hay una serie de métodos catacterísticos que nos permiten operar con ellas: añadir valores, quitarlos, indexado, filtrado, etc... En [este enlace](https://www.w3schools.com/python/python_ref_list.asp) puedes encontrar todos los métodos que podrás usar con listas.

```python
asignaturas = ["quimica", "fisica", "matematicas", "deporte"]

# Añadimos elementos a la lista con APPEND

asignaturas.append("literatura")
print(asignaturas)

# Insertamos elementos a la lista en una posición determinada
asignaturas.insert(1, "filosofia")
print(asignaturas)

# Limpiar la lista completa y dejarla vacía con CLEAR
#asignaturas.clear()
# print(asignaturas)

#Quitar un elemento que está en una posición determinada con POP. OJO: si ejecuto solo POP, me arroja el elemento de la posición indicada solito
asignaturas.pop(1)
print(asignaturas)

#Quitar un elemento por su valor con REMOVE
asignaturas.remove("deporte")
print(asignaturas)

#Ordenar la lista por orden alfabético
asignaturas.sort()
print(asignaturas)

```
